```python
import random
import pandas as pd
```

```python
def roll_to_hit():
    roll = random.randint(1,6)
    if roll == 6:
        roll2 = random.randint(1,6)
        if roll2 >= 5:
            result = ["HIT", "HIT"]
        else:
            result = ["HIT", "MISS"]
    elif roll >= 5:
        result = ["HIT"]
    else:
        result = ["MISS"]
    return(result)
```

```python
roll1 = roll_to_hit()
roll1
```

```python
number_of_observations = 100000
number_of_shots = 4
observation_list = []
Deathskullz = "Y"
for x in range(number_of_observations):
    shot_count = []
    for n in range(number_of_shots):
        hit_roll = roll_to_hit()
        shot_count.extend(hit_roll)
    if Deathskullz == "Y":
        if "MISS" in shot_count:
            reroll = roll_to_hit()
            shot_count.extend(reroll)
    observation_list.append(shot_count.count("HIT"))
print(observation_list)
```
