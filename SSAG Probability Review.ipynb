```python
import random
import pandas as pd
```

```python
def roll_to_hit():
    roll = random.randint(1,6)
    if roll == 6:
        roll2 = random.randint(1,6)
        if roll2 >= 5:
            result = ["HIT", "HIT"]
        else:
            result = ["HIT", "MISS"]
    elif roll >= 5:
        result = ["HIT"]
    else:
        result = ["MISS"]
    return(result)
```

```python
Toughness = 8
def roll_to_wound(Toughness, strength):
    if Toughness >= 2*strength:
        target = 6
    elif Toughness > strength:
        target = 5
    elif Toughness == strength:
        target = 4
    elif Toughness < strength/2:
        target = 2
    elif Toughness < strength:
        target = 1
    roll = random.randint(1,6)
    if roll >= target:
        result = ["WOUND"]
    else:
        result = ["FAIL"]
    return(result)
```

```python
armor = 2
def save_and_damage(armor):
    ''' roll = random.randint(1,6)
    if roll < armor+3:
        damage = random.randint(1,6)
    else:
        damage = 0'''
    damage = random.randint(1,6)
    return(damage)
```

```python
number_of_observations = 10000
#number_of_shots = 3
observation_list = []
Deathskullz = "Y"
for x in range(number_of_observations):
    shot_count = []
    shots1 = random.randint(1,6)
    shots2 = random.randint(1,6)
    number_of_shots = shots1+shots2
    strength1 = random.randint(1,6)
    strength2 = random.randint(1,6)
    strength = strength1+strength2
    for n in range(number_of_shots):
        hit_roll = roll_to_hit()
        shot_count.extend(hit_roll)
    if Deathskullz == "Y":
        if "MISS" in shot_count:
            reroll = roll_to_hit()
            shot_count.extend(reroll)
    observation_list.append([shot_count.count("HIT"), strength])
print(observation_list)
```

```python
wound_observation_list = []
for x in range(len(observation_list)):
    wound_count = []
    strength = observation_list[x][1]
    mortal = 0
    if strength > 10:
        for n in range (observation_list[x][0])
        mortal = mortal + random.randint(1,3)
    for n in range(observation_list[x][0]):
        wound_roll = roll_to_wound(Toughness, strength)
        #print(wound_roll)
        wound_count.extend(wound_roll)
        #print(wound_count)
    if Deathskullz == "Y":
        if "FAIL" in wound_count:
            reroll = roll_to_wound(Toughness, strength)
            #print(reroll)
            wound_count.extend(reroll)
            #print(wound_count)
    wound_observation_list.append(wound_count.count("WOUND") + mortal)
print(wound_observation_list)            
```

```python
damage_observation_list = []
for x in range (len(wound_observation_list)):
    damage_count = []
    for n in range(wound_observation_list[x]):
        damage_roll = save_and_damage(armor)
        damage_count.append(damage_roll)
    if Deathskullz == "Y":
        if 1 in damage_count:
            damage_count.remove(1)
            reroll = save_and_damage(armor)
            damage_count.append(reroll)
        elif 2 in damage_count:
            damage_count.remove(2)
            reroll = save_and_damage(armor)
            damage_count.append(reroll)
    damage_observation_list.append(sum(damage_count))
print(damage_observation_list)
```
